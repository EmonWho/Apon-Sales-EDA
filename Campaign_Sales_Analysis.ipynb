```python
# Assuming your CSV file is named 'your_dataset.csv' and it's uploaded to the root of your Google Drive
file_path = '//content/drive/MyDrive/Dataset/Data for Assessment/Sales Data.csv'

# Load the dataset
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics
summary_stats = data.describe()
```

```python
# Visualize the distribution of numerical features
sns.pairplot(data)
plt.show()

# Correlation heatmap
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Outlet', y='Total Discount', data=data, alpha=0.5)
plt.title('Outlet vs Total Discount')
plt.xlabel('Outlet')
plt.ylabel('Total Discount')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Total Sales Amount', y='Total Discount', data=data, alpha=0.5)
plt.title('Total Sales Amount vs Total Discount')
plt.xlabel('Total Sales Amount')
plt.ylabel('Total Discount')
plt.show()
```
